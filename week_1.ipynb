# Langkah (Step) 2

Baca Image dari URLs - Read Image from URLs

--Bahasa--

Pada langkah ini kita akan membaca image dari URLs, menampilkannya menggunakan OpenCV, perhatikan ada perbedaaan keitka membuka image dalam RGB dan BGR. Channel warna default untuk OpenCV adalah BGR

--English--

in this step we will read images from uris, and display them using OpenCV, please note the difference hen reading image in RGB and BGR, and BGR format the default input color channels are in BGR format for OpenCv. RGB?

```python
# Membuat List untuk menyimpan url dari bebrapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://asset.kompas.com/crops/gnABECiqt4rInhX9U7yRXJ8-Lww=/10x0:593x389/1200x800/data/photo/2023/01/19/63c8e3f30f7ea.jpeg"]

# looppada tiap url image, bebrapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

# langkah (Step) 3:

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

# langkah (Step) 4:

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
import cv2 as cv
from skimage import io
from google.colab.patches import cv2_imshow

url = "https://asset.kompas.com/crops/gnABECiqt4rInhX9U7yRXJ8-Lww=/10x0:593x389/1200x800/data/photo/2023/01/19/63c8e3f30f7ea.jpeg"
image = io.imread(url)
image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi, lebar = image_rgb.shape[:2]
image_mod = image_rgb.copy()

# Garis horizontal putih di tengah
for x in range(lebar):
    image_mod[tinggi // 2, x] = [255, 255, 255]

cv2_imshow(image_mod)
```

```python
image_mod = image_rgb.copy()

# Garis vertikal putih di tengah
for y in range(tinggi):
    image_mod[y, lebar // 2] = [255, 255, 255]

cv2_imshow(image_mod)
```

```python
image_mod = image_rgb.copy()

# Garis diagonal putih
for i in range(min(tinggi, lebar)):
    image_mod[i, i] = [255, 255, 255]

cv2_imshow(image_mod)
```

```python
import random

image_mod = image_rgb.copy()

# Kotak putih random
box_w, box_h = 100, 60
start_x = random.randint(0, lebar - box_w)
start_y = random.randint(0, tinggi - box_h)

image_mod[start_y:start_y+box_h, start_x:start_x+box_w] = [255, 255, 255]

cv2_imshow(image_mod)
print(f"Letak kotak: x={start_x}, y={start_y}")
```
